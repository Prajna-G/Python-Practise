```python
import matplotlib.pyplot as plt
import numpy as np

def plot_heart(x, y, ratio):
    fig = plt.figure(figsize=(4,4), facecolor="black")
    ax = plt.gca()
    ax.set_facecolor("black")
    x = x * np.sin(ratio)
    y = y * np.sin(ratio)
    ax.scatter(x, y, s=3, c="#D91E2A")
    plot_scatter(ax, x, y, 0.15, 1)
    plot_scatter(ax, x, y, 0.15, 2) 
    plot_scatter(ax, x, y, 0.15, 3)
    xi = x[0:x.shape[0]:2] * np.sin(ratio) * .7
    yi = x[0:y.shape[0]:2] * np.sin(ratio) * .7
    plot_scatter(ax, xi, yi, 0.25, 4)
    xo = x[0:x.shape[0]:2] * np.sin(ratio) * 1.2
    yo = x[0:y.shape[0]:2] * np.sin(ratio) * 1.2
    np.random.seed(np.random.randint(1, 3))
    plot_scatter(ax, xo, yo, 0.1, 4, alpha=0.8)

    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.tick_params(bottom=False, labelbottom=False, left=False, labelleft=False)
    plt.xlim((-6*np.pi, 6*np.pi))
    plt.ylim((-6*np.pi, 6*np.pi))
    plt.tight_layout(pad=0)
    return ax

def plot_scatter(ax, x, y, size, seed, **kwargs):
    np.random.seed(seed)
    n = len(x)
    colors = np.random.rand(n)
    ax.scatter(x, y, s=size, c=colors, **kwargs)

frames = []
for ratio in np.linspace(np.pi/3, 2*np.pi/3, 1):
    t = np.linspace(0, 2*np.pi, 2000)
    x = 16 * np.sin(t) ** 3
    y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 *  np.cos(3 * t) - np.cos(4 * t)
    ax = plot_heart(x, y, ratio)
    frames.append([ax])

# Print only the first frame
frames[0][0].figure.savefig('heart_plot.png')
plt.show()
```
