```python
# CODE FOR CONVERTING AN IMAGE TO SKETCH CONVERSION
import cv2
image=cv2.imread(r'V:\ALL PROJECTS\PYTHON PROJECTS\Python Scripts\Python project\Python projects\IMAGE-1.jpg')

grey_filter=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imwrite(r'V:\ALL PROJECTS\PYTHON PROJECTS\Python Scripts\Python project\Python projects\grey_filter.png',grey_filter)
invert=cv2.bitwise_not(grey_filter)
cv2.imwrite(r'V:\ALL PROJECTS\PYTHON PROJECTS\Python Scripts\Python project\Python projects\invert.png' ,invert)
blur=cv2.GaussianBlur(invert,(21,21),-15)
cv2.imwrite(r'V:\ALL PROJECTS\PYTHON PROJECTS\Python Scripts\Python project\Python projects\blur.png',blur)
invertedblur=cv2.bitwise_not(blur)
cv2.imwrite(r'V:\ALL PROJECTS\PYTHON PROJECTS\Python Scripts\Python project\Python projects\invertedblur.png',invertedblur)
sketch=cv2.divide(grey_filter,invertedblur,scale=256.0)

cv2.imwrite(r'V:\ALL PROJECTS\PYTHON PROJECTS\Python Scripts\Python project\Python projects\CONVERTED-SKETCH.jpg',sketch)
```
