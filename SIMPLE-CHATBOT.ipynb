```python
import tkinter as tk
import webbrowser
import pygame.mixer as mixer       

# Initialize mixer
mixer.init()

# URLs
youtube_url = "https://www.youtube.com"
youtube_url2 = "https://www.google.com/search?q=pac+Man"

# Load a song
mixer.music.load(r"V:\ALL PROJECTS\PYTHON PROJECTS\Python Scripts\Python project\Songs\_Dynamite(PagalWorld.com.se).mp3")

# Available chatbot commands
chatbot_hints = """
üîπ Available Commands:
  - hi
  - how are you
  - good
  - sad
  - tell me a joke
  - what is your name
  - sing me a song
  - stop
  - i am bored
  - bye
  - ha ha
"""

def get_response():
    user_message = user_input.get().strip().lower()
    chat_display.config(state=tk.NORMAL)
    chat_display.insert(tk.END, f"You: {user_message}\n")

    # Chatbot responses
    if user_message == "hi":
        response = "Hello! Long time no see, buddy..."
    elif user_message == "how are you":
        response = "I am fine, how are you?"
    elif user_message == "good":
        response = "Hmm... How was your day by the way?"
    elif user_message == "sad":
        response = "It's okay, everything will be fine."
    elif user_message == "tell me a joke":
        response = "Why doesn't shrimp share its treasure? Because it is shellfish! üòÇüòÇüòÇ"
    elif user_message == "what is your name":
        response = "My name is Bangtanü´∞üèª"
    elif user_message == "sing me a song":
        mixer.music.play()
        response = "Playing song..."
    elif user_message == "stop":
        mixer.music.stop()
        response = "Music stopped."
    elif user_message == "i am bored":
        webbrowser.open(youtube_url2)
        response = "Opening something fun for you..."
    elif user_message == "bye":
        response = "Goodbye! Have a great day."
    elif user_message == "ha ha":
        response = "Ha ha ha..."
    else:
        response = "Sorry, I didn't catch that."

    chat_display.insert(tk.END, f"Chatbot: {response}\n\n")
    chat_display.config(state=tk.DISABLED)
    user_input.delete(0, tk.END)

# Create main window
window = tk.Tk()
window.configure(bg="light gray")
window.title("Bangtanüíú")

# Create chat display with scrollbar
chat_frame = tk.Frame(window)
chat_scroll = tk.Scrollbar(chat_frame)
chat_display = tk.Text(chat_frame, height=15, width=60, bg="lavender", state=tk.DISABLED, yscrollcommand=chat_scroll.set)
chat_scroll.pack(side=tk.RIGHT, fill=tk.Y)
chat_display.pack(side=tk.LEFT, fill=tk.BOTH)
chat_scroll.config(command=chat_display.yview)
chat_frame.pack(pady=5)

# Create user input field
user_input = tk.Entry(window, width=75, bg="white")
user_input.pack(pady=5)

# Create send button
send_button = tk.Button(window, text="Send", command=get_response, bg="grey")
send_button.pack(pady=5)

# Hint Section
hint_label = tk.Label(window, text=chatbot_hints, justify="left", bg="light gray", fg="black", font=("Arial", 10))
hint_label.pack(pady=5)

# Start Tkinter loop
window.mainloop()
```
